```python
from obspy import read, read_events
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Stack stations
flist = glob("matrix_profiles/20200201000000_20200204000000*.mp")

stack = np.loadtxt(flist[0])
for file in flist[1:]:
    print(file)
    tmp = np.loadtxt(file)
    fig, ax = plt.subplots(figsize=(15,15))
    ax.plot(tmp)
    plt.close()
    stack += tmp
```
